Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection & Processing

```python
# loading the data from csv file to a Pandas DataFrame
calories = pd.read_csv('/content/calories.csv')
```

```python
# print the first 5 rows of the dataframe
calories.head()
```

```python
exercise_data = pd.read_csv('/content/exercise.csv')
```

```python
exercise_data.head()
```

Combining the two Dataframes

```python
calories_data = pd.concat([exercise_data, calories['Calories']], axis=1)
```

```python
calories_data.head()
```

```python
# checking the number of rows and columns
calories_data.shape
```

```python
# getting some informations about the data
calories_data.info()
```

```python
# checking for missing values
calories_data.isnull().sum()
```

Data Analysis

```python
# get some statistical measures about the data
calories_data.describe()
```

Data Visualization

```python
sns.set()
```

```python
# plotting the gender column in count plot
sns.countplot(calories_data['Gender'])
```

```python
# finding the distribution of "Age" column
sns.distplot(calories_data['Age'])
```

```python
# finding the distribution of "Height" column
sns.distplot(calories_data['Height'])
```

```python
# finding the distribution of "Weight" column
sns.distplot(calories_data['Weight'])
```

```python
# Pairplot for visualizing the relationships
plt.figure(figsize=(12, 8))
sns.pairplot(calories_data[['Duration', 'Age', 'Weight', 'Height', 'Calories', 'Heart_Rate', 'Body_Temp']])
plt.title('Pairplot of Key Features')
plt.show()
```

```python
calories_data['Gender'] = calories_data['Gender'].map({'male': 0, 'female': 1})
```

```python
calories_data.head()
```

```python
calories_data.to_csv('extracted_data.csv', index=False)
print("Data successfully saved as a CSV file!")
```

```python
# Distribution of calories burned
plt.figure(figsize=(10, 6))
sns.histplot(calories_data['Calories'], kde=True, color='blue')
plt.title('Distribution of Calories Burned')
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.show()
```

Finding the Correlation in the dataset

```python
correlation = calories_data.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
